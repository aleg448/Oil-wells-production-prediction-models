# Regression models

**Multiple linear Regression models**

SGD Regressor, BayesianRidge, LassoLars, ARDRegression, passiveAggressiveRegressor, TheilSenRegressor, LinearRegression

```python
import numpy as np
from sklearn import linear_model
from sklearn import svm

classifiers = [
    svm.SVR(),
    linear_model.SGDRegressor(),
    linear_model.BayesianRidge(),
    linear_model.LassoLars(),
    linear_model.ARDRegression(),
    linear_model.PassiveAggressiveRegressor(),
    linear_model.TheilSenRegressor(),
    linear_model.LinearRegression()]


train_features, test_features, train_labels, test_labels = train_test_split(purgedData3.drop(columns=["production"]),purgedData3['production'], test_size = 0.2, random_state = 31)

for item in classifiers:
    print(item)
    clf = item
    clf.fit(train_features, train_labels)
    predictions=clf.predict(test_features)
    errors = abs(predictions - test_labels)
    print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
    mape = 100 * (errors / test_labels)
    accuracy = 100 - np.mean(mape)
    print('Accuracy:', round(accuracy, 2), '%.')
```

# Decision tree model

```python

from sklearn import tree
from sklearn import preprocessing
from sklearn import utils

purgedData3Int=purgedData3.astype('int')

# Split data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(purgedData3Int.drop(columns=["production"]),purgedData3Int['production'], test_size = 0.2, random_state = 1)

# create regressor object
regressor = tree.DecisionTreeClassifier()
 
# fit the regressor with x and y data
regressor=regressor.fit(train_features, train_labels) 

# Use the forest's predict method on the test data
predictions = regressor.predict(test_features)

# Calculate the absolute errors
errors = abs(predictions - test_labels)

# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')

# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)

# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')

```
